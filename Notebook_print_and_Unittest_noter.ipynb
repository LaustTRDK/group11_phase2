# Unittest

For at køre unitest så skal jeg lave et modul som hedder noget med test hvor alle unittest metoderne vil være inde under. 

    Stat dokumentet med at tilføre:
        import unittest
        import random
        from you_code_file_you_want_to_test import those_classes_you_want_to_test

Derefter er det at lave klasser der passer med hver klasse fra dit dokument du gerne vil teste. Under disse klasser skal der stå 'metods' der er for hver enkelt test du gerne lave. 
Docktest laver det man kalder for en 'happy' test. Den tester hvos noget virker. Men unnittest så har man også mulighed for at teste om noget fejler ved / med de betingelser koden burde fejle med. 

For at køre ens unittest dokument med ens test inde i skal man i terminalen skrive: 
    python -m unittest you_test_file.py

fx i et af mine projekter så mens jeg står i den mappe hvor unittest filen er inde i kørte jeg en test og min fil hed 'unittest.py'. 
    python -m unittest unittest.py

Der er mange måder at skrive testkode på og mange af de methods der for at vurdere kan man finde via: 
https://www.geeksforgeeks.org/python/unit-testing-python-unittest/

Det er vigtigt at du står i den mappe hvori din unitestfil er i samt også den fil som du tester igennem din testfil. 

Slutningen af filen skal have: 
    
    if __name__ == '__main__':
    unittest.main()
